#### Row Filtering

```python
df_filtered = df.filter((df.release_year >= 2000) & (df.release_year <= 2010))
```

```python
from pyspark.sql.functions import col
df_filtered = df.filter((col("release_year") >= 2000) & (col("release_year") <= 2010))
df_filtered.display()
```

```python
df.filter(col('release_year').between(2000,2010)).display()
```

```python
df.filter(col('studio') == 'Marvel Studios').display()
```

```python
df.select("industry").distinct().display()
```

```python
df.select("language").distinct().display()
```

```python
df = df.withColumn('profit', col('revenue') - col('budget'))
df.display()
```

```python
df = df.withColumnRenamed('revenue', 'total__revenue')
df.printSchema()
```
